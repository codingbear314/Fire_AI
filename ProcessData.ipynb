```python
Dataset = open('../Dataset/Lichess_200GB.pgn', 'r')
Trim_Dataset = open('../Dataset/Lichess_Trimmed.txt', 'w')
```

```python
k = 0
for i in Dataset:
    i = i.strip()
    if i == '':
        continue
    if i[0] == '[':
        continue
    if 'eval' not in i:
        continue
    Trim_Dataset.write(i + '\n')
    if k%100000 == 0:
        print(k)
        print(i)
    k += 1
```

```python
Trim_Dataset.close()
```
